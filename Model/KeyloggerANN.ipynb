```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./FinalKeyloggerCustom.csv")
df
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(['index',"Unnamed: 0","fileinfo"],axis="columns",inplace=True)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df['class'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_scale=['e_maxalloc','E_file',"filesize","e_sp","SizeOfCode","SizeOfInitializedData","SizeOfUninitializedData","AddressOfEntryPoint",
              "BaseOfCode","BaseOfData","MajorImageVersion","CheckSum","SizeOfStackReserve","SizeOfStackCommit","SizeOfHeapReserve","SizeOfHeapCommit","non_sus_sections"
              ,"sus_sections"]
```

```python
df[cols_to_scale]=scaler.fit_transform(df[cols_to_scale])
df['filesize']
```

```python
x=df.drop(['packer_type',"OH_DLLchar9",],axis="columns")
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
x_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(48,input_shape=(67,),activation='sigmoid'),
    keras.layers.Dense(17,input_shape=(48,),activation='sigmoid'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=10,validation_split=0.2)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp=model.predict(x_test)
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
